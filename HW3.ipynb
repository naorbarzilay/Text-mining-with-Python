```python
# using nlp to find the pos of the words in the tweetsand add the words to the correct place (verb/adj/noun)
nlp = spacy.load("en_core_web_sm")

def words_and_POS(text):
    english_vocab = sorted(set(w.lower() for w in nltk.corpus.words.words()))
    stopwords = nltk.corpus.stopwords.words('english')
    lemmatizer = WordNetLemmatizer()
    verbs = []
    adjs = []
    nouns = []
    splited_text = text.split('\n')
    for sen in splited_text:
        sen=sen.lower()
        sen = sen.lstrip().rstrip()
        doc = nlp(sen)
        for token in doc:
            if(token.text not in stopwords and token.text in english_vocab):
                if(token.pos_== "VERB"):
                    verbs.append(lemmatizer.lemmatize(token.text , pos= "v"))  
                if(token.pos_== "ADJ"):
                    adjs.append(lemmatizer.lemmatize(token.text , pos= "a")) 
                if(token.pos_== "NOUN"):
                    nouns.append(lemmatizer.lemmatize(token.text , pos= "n")) 
    return [verbs,adjs,nouns] # [0] = verbs , [1] = adj , [2] = nouns for any text that the function got

list_pos_aids = words_and_POS(aids_text)
list_pos_cancer = words_and_POS(cancer_text)
list_pos_covid = words_and_POS(covid_data)
list_pos_diabetes = words_and_POS(diabetes_text)
```

## Q3 Parsing

#### A - Verbs

```python
# finds all the verbs that were connected to the disease
def verb_dependency(text, disease):
    nlp = spacy.load("en_core_web_sm")
    splited_text = text.split('\n')
    verbs=[]
    lemmatizer = WordNetLemmatizer()
    for tweet in splited_text:
        tweet = tweet.lstrip().rstrip() # remove extra white spaces (beginning and end of tweets)
        doc = nlp(tweet)
        for token in doc:
            word = token.text.lower()
            if(word==disease and token.head.pos_=="VERB"):
                verbs.append(lemmatizer.lemmatize(token.head.text.lower(), "v"))
    return verbs

list_aids = verb_dependency(aids_text, "aids")
list_cancer = verb_dependency(cancer_text, "cancer")
list_covid = verb_dependency(covid_data, "covid")
list_diabetes = verb_dependency(diabetes_text, "diabetes")
```

```python
# finds all the adjectives that were connected to the disease

def adjective_dependency(text, disease):
    nlp = spacy.load("en_core_web_sm")
    splited_text = text.split('\n')
    adjectives=[]
    lemmatizer = WordNetLemmatizer()
    for tweet in splited_text:
        tweet = tweet.lstrip().rstrip() # remove extra white spaces (beginning and end of tweets)
        doc = nlp(tweet)
        for token in doc:
            word = token.head.text.lower()
            if(token.pos_=="ADJ" and word==disease):
                adjectives.append(lemmatizer.lemmatize(token.text.lower(),"a"))
    return adjectives


list_aids = adjective_dependency(aids_text, "aids")
list_cancer = adjective_dependency(cancer_text, "cancer")
list_covid = adjective_dependency(covid_data, "covid")
list_diabetes = adjective_dependency(diabetes_text, "diabetes")
```

```python
# We relaized that the vocabulary had words that had no meanings like "k" so we only wanted words that has more than 2 letters
english_vocab = sorted(set(w.lower() for w in nltk.corpus.words.words() if len(w)>2))
```

```python
# Cleaing the text
# Building a list that has lists of texts- each internal list has the vocabulary of that text
all_train_words = [] 

for text in news_group_data:
    processed_article = text.lower()
    processed_article = re.sub('[^a-zA-Z]', ' ', processed_article) # replace anything that not a word to ' '
    processed_article = re.sub(r'\s+', ' ', processed_article) # Eliminate duplicate whitespaces using wildcards

    # Preparing the dataset
    all_sentences = nltk.sent_tokenize(processed_article)

    all_words = [nltk.word_tokenize(sent) for sent in all_sentences] # all the words from the current text
    # Removing Stop Words and not vaild words
    for i in range(len(all_words)):
        all_words[i] = [w for w in all_words[i] if w not in stopwords.words('english') and w in english_vocab]
    all_train_words += all_words
```

```python
all_test_words = [] 

for text in news_group_test.data:
    processed_article = text.lower()
    processed_article = re.sub('[^a-zA-Z]', ' ', processed_article)
    processed_article = re.sub(r'\s+', ' ', processed_article)
    # Preparing the dataset
    all_sentences = nltk.sent_tokenize(processed_article)

    all_words = [nltk.word_tokenize(sent) for sent in all_sentences]

    # Removing Stop Words
    for i in range(len(all_words)):
        all_words[i] = [w for w in all_words[i] if w in english_vocab and w not in stopwords.words('english')]
    all_test_words += all_words
```

```python
all_corpus_words = all_test_words + all_train_words # text and train data
all_corpus_words
```

```python
start_time = time.time()

word2vec = Word2Vec(all_corpus_words, min_count=2,window =2)
vocabulary = word2vec.wv.vocab
vocabulary
```

```python
def get_mean_vector(word2vec_model, words):
    # remove out-of-vocabulary words
    words = [word for word in words if word in word2vec_model.wv.vocab]
    if len(words) >= 1:
        return np.mean(word2vec_model[words], axis=0)
    else:
        return []

train_vectors = []
for train_text_words in all_train_words:
    train_vectors.append(get_mean_vector(word2vec,train_text_words))
    
test_vectors = []
for test_text_words in all_test_words:
    test_vectors.append(get_mean_vector(word2vec,test_text_words))    

cosine_similarity(train_vectors, test_vectors, dense_output=True)

seconds = time.time() - start_time
print('Time Taken:', time.strftime("%H:%M:%S",time.gmtime(seconds)))
```

```python
df = pd.DataFrame(cosine_similarity(train_vectors, test_vectors, dense_output=True))
df
```

```python
cos_predict = df.idxmax(axis = 0)
cos_predict
```

```python
# find the topic using the topic of the closet neighbor according to the cosine similarity

new_df_word2vec = pd.DataFrame(columns=['similarity_text', 'per_group'])

for i in range(len(cos_predict.values)):
    row_df = pd.DataFrame({'similarity_text': [cos_predict.values[i]],'per_group': [news_group_target_names[news_group_target[cos_predict.values[i]]]]})
    new_df_word2vec = new_df_word2vec.append(row_df,ignore_index=True)
new_df_word2vec 
```

```python
cos_predict = cos_predict.values
cos_predict = [news_group_target[i] for i in cos_predict]

accuracy_score(news_group_test.target, cos_predict)
```
